```python
import pandas as pd

att = pd.read_csv("data\\attorneys.csv")
ate = pd.read_csv("data\\attorneytimeentries.csv")
att.head(5)
ate.head(5)
```

```python
attorneys = pd.DataFrame(att.StateAbbr.value_counts()).rename(columns={"StateAbbr": "Attorneys"})
attorneys
```

```python
cli = pd.read_csv("data\\clients.csv")
clients = pd.DataFrame(cli.StateAbbr.value_counts()).rename(columns={"StateAbbr": "Clients"})
clients
```

```python
subcat = pd.read_csv("data\\subcategories.csv")
subcatcounts = subcat.groupby(['StateAbbr', 'Subcategory']).count()
subcatcounts.to_csv('subcatcounts.csv')
subcatcounts
```

```python
attcli = pd.merge(attorneys, clients, left_index=True, right_index=True).reset_index()
attcli['Ratio'] = attcli.Attorneys/attcli.Clients
attcli.to_csv('attorneysandclients.csv')
attcli
```
